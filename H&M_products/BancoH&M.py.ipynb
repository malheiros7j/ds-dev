```python
import sqlite3
import pandas as pd

#Ler o dataframe

data = pd.read_csv('products_H&M_Final.csv')

query_showroom_schema = """
	
	CREATE TABLE showroom(
		product_id				TEXT,
		fit						TEXT,
		product_name			TEXT,
		product_price			REAL,
		color_name				TEXT,	
		style_id				INTEGER,
		color_id				INTEGER,
		scrapy_datetime			TEXT,
		size_number				INTEGER,
		size_model				TEXT,
		cotton					REAL,
		polyester				REAL,
		elastane				REAL,
		elasterell				REAL)"""
```

```python
# Connect do DB
conn = sqlite3.connect('h&m_db.sqlite')
cursor = conn.execute(query_showroom_schema)
conn.commit()
conn.close()
```
