```python
import sqlite3
import pandas as pd
from sqlalchemy import create_engine
```

```python
db = create_engine('sqlite:///db_olist.sqlite',echo=False)
conn = db.connect()
```

```python
## DATA READING
df_customer = pd.read_csv('DataBases/olist_customers_dataset.csv')
df_geolocation = pd.read_csv('DataBases/olist_geolocation_dataset.csv')
df_order_items = pd.read_csv('DataBases/olist_order_items_dataset.csv')
df_order_payments = pd.read_csv('DataBases/olist_order_payments_dataset.csv')
df_order_reviews = pd.read_csv('DataBases/olist_order_reviews_dataset.csv')
df_orders = pd.read_csv('DataBases/olist_orders_dataset.csv')
df_products = pd.read_csv('DataBases/olist_products_dataset.csv')
df_sellers = pd.read_csv('DataBases/olist_sellers_dataset.csv')
df_product_category_name = pd.read_csv('DataBases/product_category_name_translation.csv')
```

```python
schema_customer = """
    CREATE TABLE customer(
    customer_id             text,
    customer_unique_id      text,
    customer_zip_code_prefix integer,
    customer_city           text,
    customer_state          text)

"""
schema_geolocation ="""
    CREATE TABLE geolocation(
    geolocation_zip_code_prefix    integer,
    geolocation_lat                real,
    geolocation_lng                real,
    geolocation_city               text,
    geolocation_state              text)

"""
schema_order_items = """
    CREATE TABLE order_items(
    order_id                text,
    order_item_id           integer,
    product_id              text,
    seller_id               text,
    shipping_limit_date     text,
    price                   real,
    freight_value           real)

"""
schema_order_payments = """
    CREATE TABLE order_payments(
    order_id                 text,
    payment_sequential       integer,
    payment_type             text,
    payment_installments     integer,
    payment_value            real)

"""
schema_order_reviews = """
    CREATE TABLE order_reviews(
    review_id                  text,
    order_id                   text,
    review_score                integer,
    review_comment_title       text,
    review_comment_message     text,
    review_creation_date       text,
    review_answer_timestamp    text)

"""
schema_orders = """
    CREATE TABLE orders(
    order_id                         text,
    customer_id                      text,
    order_status                     text,
    order_purchase_timestamp         text,
    order_approved_at                text,
    order_delivered_carrier_date     text,
    order_delivered_customer_date    text,
    order_estimated_delivery_date    text)

"""
schema_products = """
    CREATE TABLE products(
    product_id                     text,
    product_category_name          text,
    product_name_lenght           real,
    product_description_lenght    real,
    product_photos_qty            real,
    product_weight_g              real,
    product_length_cm             real,
    product_height_cm             real,
    product_width_cm              real)

"""
schema_sellers = """
    CREATE TABLE sellers(
    seller_id                 text,
    seller_zip_code_prefix    integer,
    seller_city               text,
    seller_state              text)

"""
schema_product_category_name = """
    CREATE TABLE product_category_name(
    
    product_category_name            text,
    product_category_name_english    text)

"""


# Create schema
conn.execute(schema_customer)
conn.execute(schema_geolocation)
conn.execute(schema_order_items)
conn.execute(schema_order_payments)
conn.execute(schema_order_reviews)
conn.execute(schema_orders)
conn.execute(schema_products)
conn.execute(schema_sellers)
conn.execute(schema_product_category_name)


# Insert data into table - Index = False para resetar no começo
df_customer.to_sql('customer',con=conn,if_exists='append',index=False)
df_geolocation.to_sql('geolocation',con=conn,if_exists='append',index=False)
df_order_items.to_sql('order_items',con=conn,if_exists='append',index=False)
df_order_payments.to_sql('order_payments',con=conn,if_exists='append',index=False)
df_order_reviews.to_sql('order_reviews',con=conn,if_exists='append',index=False)
df_orders.to_sql('orders',con=conn,if_exists='append',index=False)
df_products.to_sql('products',con=conn,if_exists='append',index=False)
df_sellers.to_sql('sellers',con=conn,if_exists='append',index=False)
df_product_category_name.to_sql('product_category_name',con=conn,if_exists='append',index=False)
```
